# 지도학습(Supervised Learning)
*비지도학습과의 차이점: 지도학습은 입력데이터와 출력데이터가 존재하나 비지도학습은 입력데이터만 존재*

- 사람과 사물 분류
- 스팸 메일 분류
- 지문, 홍채 인식 판별
2. 회귀(Regression)

- 주식 가격 예측

< 딥러닝 학습 순서>
1. imoort: 필요한 모듈(라이브러리)를 import
2. 전처리: 학습에 필요한 데이터 전처리를 수행

4. 컴파일: 모델을 생성

```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([5.0, 6.0, 7.0, 8.0, 9.0, 10.0], dtype=float)

model = Sequential([
    Dense(3, input_shape=[1]),
    Dense(4),
    Dense(4),
    Dense(1),
])
# input_shape[1]도 되고
# input_shape(1)도 된다
# 1인 이유는 데이터가 하나씩 들어가기 때문이다.
```

```python
model = Sequential([
    Dense(1, input_shape=[1]),
])
# 단순선형회귀이므로 단순한 모델링
```

```python
model.compile(optimizer='sgd', loss='mse')
# sgd: Stochastic Gradient Descent
```

```python
# fit 메소드를 사용하여 학습
model.fit(xs, ys, epochs=1200, verbose=0)
```

```python
# 예측: 학습이 잘 되었는지 확인
model.predict([10.0])
```
